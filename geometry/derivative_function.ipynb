```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function definition

# Example with the Gaussian function
def gauss_funk(mu, sigma, x):
    coef = 1/((2*np.pi*sigma**2)**0.5)
    exp_arg = -((x-mu)**2)/(2*sigma**2)
    return coef*np.exp(exp_arg)

# Derivative function
def derive_funk(x, y, order):
    _coord = list(zip(x, y))
    out = [None for i in range(0, order)]
    
    for i in range(order, len(_coord)):
        x_dist = _coord[i][0] - _coord[i-1][0]
        y_dist = _coord[i][1] - _coord[i-1][1]
        slope = y_dist/x_dist
        out.append(slope)
    return out
```

```python
# Run the functions and plots
# Parameters
MU = 0.0
SIGMA = 1.0
SIZE = 100
WIDTH = 3

coord_x = np.linspace(WIDTH*(-1), WIDTH, SIZE)

#Gaussian coordinates
gauss_y = gauss_funk(MU, SIGMA, coord_x)

# 1st derivative coordinates
deriv1_y = derive_funk(coord_x, gauss_y, 1)

# 2nd derivative coordinates
deriv2_y = derive_funk(coord_x, deriv1_y, 2)

# 3rd derivative coordinates
deriv3_y = derive_funk(coord_x, deriv2_y, 3)

# Illustration
plt.plot(coord_x, gauss_y, '#373A36', label='Gaussian function')
plt.plot(coord_x, deriv1_y, '#F1567B', label='1st derivative')
plt.plot(coord_x, deriv2_y, '#00A499', label='2nd derivative')
plt.plot(coord_x, deriv3_y, '#FFD400', label='3rd derivative')
plt.legend(loc='best')
```
